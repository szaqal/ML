```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


%matplotlib inline
```

### Seaborn example

```python
plt.figure(figsize=(3,1))
x = np.linspace(0,10,100)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus)
plt.show()

plt.figure(figsize=(3,1))
plt.plot(x, sinus, "o") # dotted
plt.show()

plt.figure(figsize=(3,1))
plt.plot(x, sinus, "-b", x, sinus, "ob", x, cosinus, "-r", x, cosinus, "or")
plt.show()
```

### Pandas dataset plot

[BOX plot](https://miroslawmamczur.pl/003-wykres-pudelkowy-box-plot/)

```python
salary = pd.read_csv("https://github.com/duchesnay/pystatsml/raw/master/datasets/salary_table.csv")
df = salary
print(df.head())
plt.figure(figsize=(3,3))
_ = sns.scatterplot(x="experience", y="salary", hue="education", data=salary)
plt.figure(figsize=(3,3))
_ = sns.lineplot(x="experience", y="salary", hue="education", data=salary)

plt.figure(figsize=(3,3))
_ = sns.boxplot(x="management", y="salary", data=salary)


plt.figure(figsize=(3,3))
_ = sns.displot(x="salary", hue="management", kind="kde", data=salary, fill=True)
```
